Creating 3 layer neural network:
- 784 input neurons
- 30 neurons in hidden layer
- 10 in output layer

```python
net = Network.Network([784, 30, 10])
```

Training

```python
%time net.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```

```python
error = 1 - np.array(net.error) / 10000
```

```python
plt.plot(error)
plt.ylabel('error')
plt.xlabel('epoch')
plt.show()
```
